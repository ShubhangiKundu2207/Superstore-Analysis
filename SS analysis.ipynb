```python
px.line(data_frame=df,x='Order Date',y='Sales',title='Time Analysis of Sales')
```

```python
px.line(data_frame=df,x='Order Date',y='Profit',title='Time Analysis of Profit')
```

```python
fig = px.line(data_frame=df, x='Order Date', y='Sales', color_discrete_sequence=['blue'])

fig.add_trace(
    px.line(data_frame=df, x='Order Date', y='Profit',  color_discrete_sequence=['red']).data[0]
)

fig.data[1].update(showlegend=True,name='Profit')
fig.data[0].update(showlegend=True,name='Sales')
fig.update_yaxes(title_text='Value')
fig.update_layout(title='Sales and Profit Time Analysis')
fig.show()
```

```python
px.pie(data_frame=df,values='Sales',names='Category',title='Number of Orders for each Category')
```

```python
df_category_profit_sales= df.groupby('Category')[['Profit','Sales']].sum().reset_index()
```

```python
fig = px.bar(df_category_profit_sales, x='Category', y=['Profit', 'Sales'], barmode='group')
fig.update_layout(
    xaxis_title='Category',
    yaxis_title='Value',
    title='Profit and Sales by Category',
)
```

```python
region_orders =df['Region'].value_counts()
px.bar(region_orders,title='Number of Orders for each region')
```

```python
df_states_profit_sales = df.groupby('Region')[['Profit', 'Sales']].sum().reset_index()

fig = px.bar(df_states_profit_sales, x='Region', y=['Profit', 'Sales'], barmode='group')
fig.update_layout(
    xaxis_title='States',
    yaxis_title='Value',
    title='Profit and Sales by Region'
)

fig.show()
```

```python
sub_categories_order = df['Sub-Category'].value_counts()
px.bar(sub_categories_order,title='Number of Orders for each Sub Category')
```

```python
sub_categories_sales_profit = df.groupby('Sub-Category')[['Sales','Profit']].sum().reset_index()

px.bar(data_frame=sub_categories_sales_profit,x='Sub-Category',y=['Sales','Profit'],barmode ='group',title='Sales and Profit for each Sub-Category')
```

```python
px.scatter(data_frame=df, x='Sales', y='Profit', color=df['Profit'] > 0,title='Sales with +ve Profit (blue) and -ve Profit (red)')
```

```python
sns.boxplot(x=df['Category'],y=df['Profit'])
plt.title("Categorical Profit Analysis")
plt.show()
```
